```python
import random
import string

number_of_users = 9999

def generate_usernames(num):
    usernames = set()  # Use a set to avoid duplicates

    while len(usernames) < num:
        username = "".join(random.choices(string.ascii_lowercase + string.digits, k=8))
        usernames.add(username)

    return list(usernames)

# Generate N unique fake usernames
fake_usernames = generate_usernames(number_of_users)
```

```python
from faker import Faker
import pandas as pd

# Initialize a Faker generator
fake = Faker()

# Function to generate unique names using the faker library
def generate_unique_faker_names(num):
    unique_names = set()

    while len(unique_names) < num:
        # Generate a fake name
        fake_name = fake.name().split()
        # Extract first and last name assuming a two-part name
        if len(fake_name) == 2:
            first_name, last_name = fake_name[0], fake_name[1]
            unique_names.add((first_name, last_name))
        if len(fake_name) > 2:
            first_name, last_name = fake_name[1], fake_name[2]
            unique_names.add((first_name, last_name))

    # Unzip the set of tuples into two lists
    first_names, last_names = zip(*unique_names)

    return list(first_names), list(last_names)

# Generate 1000 unique first and last names
faker_first_names, faker_last_names = generate_unique_faker_names(number_of_users)

# Verify the length of the lists
len(faker_first_names), len(faker_last_names)
```

```python
import airportsdata
# Get a list of all major airports
major_airports = airportsdata.load('IATA')
random_airport_codes = random.choices(list(major_airports.keys()), k=number_of_users)
corresponding_airport_cities = [major_airports[pick]["name"] for pick in random_airport_codes]
```

```python
from datetime import datetime, timedelta
table1 = {"last_updated_timestamp": [datetime.now()-timedelta(days=5) for _ in range(len(fake_usernames))],
        "user_id": fake_usernames,
          "first_name": faker_first_names,
          "last_name": faker_last_names,
          "home_airport_code": random_airport_codes,
          "home_airport_name": corresponding_airport_cities
          }

user_home_airport = pd.DataFrame(table1)

# DEMO PURPOSES: Creates a consistent user so travel recommendations are less wildcard
new_row_data = {
    "last_updated_timestamp": datetime.now() - timedelta(days=5),
    "user_id": "demouser123",
    "first_name": "JC",
    "last_name": "Denton",
    "home_airport_code": "SFO",
    "home_airport_name": "San Francisco International Airport"
}

new_row_df = pd.DataFrame([new_row_data])
user_home_airport = pd.concat([user_home_airport, new_row_df], ignore_index=True)

user_home_airport
```

```python
import random

start_date = datetime.strptime('2023-01-01', '%Y-%m-%d')
end_date = datetime.strptime('2023-10-01', '%Y-%m-%d')

def produce_random_trips_between_dates(fake_usernames, start_date, end_date): 
    # Define the date range
    total_days = (end_date - start_date).days

    list_of_records = [] 
    for username in fake_usernames: 
        row_entry = user_home_airport[user_home_airport['user_id']==username]
        home_airport = row_entry["home_airport_code"].iloc[0]
        home_airport_name = row_entry["home_airport_name"].iloc[0]
        new_airport = home_airport
        new_airport_name = None 
        while new_airport == home_airport: #this code guarantees we pick an airport that isn't the home airport
            new_airport = random.choice(list(major_airports.keys()))
            new_airport_name = major_airports[new_airport]["name"]
        random_departure = start_date + timedelta(days=random.randrange(total_days))
        random_return = random_departure + timedelta(days=random.randint(1,30))
        list_of_records.append({"transaction_timestamp": min(datetime.now() - timedelta(days=random.randint(1,60)), 
                                                        random_departure - timedelta(days=random.randint(1,60))),
                                        "user_id": username,
                                        "departure_date": random_departure,
                                        "return_date": random_return,
                                        "origin": home_airport,
                                        "origin_name": home_airport_name,
                                        "destination": new_airport,
                                        "destination_name": new_airport_name
                                        })

    return(pd.DataFrame(list_of_records), list_of_records)

user_last_trip_taken, _ = produce_random_trips_between_dates(fake_usernames,
                                           datetime.now() - timedelta(days=180),
                                           datetime.now() - timedelta(days=30))

for i in range(4): 
    user_more_trips_taken, _ = produce_random_trips_between_dates(fake_usernames,
                                           datetime.now() - timedelta(days=180),
                                           datetime.now() - timedelta(days=30))
    user_last_trip_taken = pd.concat([user_last_trip_taken, user_more_trips_taken], ignore_index=True)


#DEMO PURPOSES
random_departure = datetime.now() - timedelta(days=180)
random_return = random_departure + timedelta(days=random.randint(1,30))
new_row_data = {"transaction_timestamp": min(datetime.now() - timedelta(days=random.randint(1,60)), 
                                                        random_departure - timedelta(days=random.randint(1,60))),
                                        "user_id": "demouser123",
                                        "departure_date": random_departure,
                                        "return_date": random_return,
                                        "origin": "SFO",
                                        "origin_name": "San Francisco International Airport",
                                        "destination": "NRT",
                                        "destination_name": "Narita International Airport"
                                        }

new_row_df = pd.DataFrame([new_row_data])
user_last_trip_taken = pd.concat([user_last_trip_taken, new_row_df], ignore_index=True)

user_last_trip_taken
    
```

```python
import tecton
from tecton import BatchSource, pandas_batch_config

@pandas_batch_config(timestamp_field="transaction_timestamp")
def transactions_function():
    return user_last_trip_taken #pandas dataframe

user_historic_trips_batch = BatchSource(
    name='user_historic_trips',
    batch_config=transactions_function
)
user_historic_trips_batch.validate()
```

```python
from tecton import PushSource
from tecton.types import String, Int64, Timestamp, Field

@pandas_batch_config(timestamp_field="transaction_timestamp")
def trip_searches_function():
    return future_trip_search #pandas dataframe

input_schema = [
    Field(name="transaction_timestamp", dtype=Timestamp),
    Field(name="user_id", dtype=String),
    Field(name="departure_date", dtype=String),
    Field(name="return_date", dtype=String),
    Field(name="destination_name", dtype=String),
    Field(name="origin_name", dtype=String),
]

user_searches_eventsource = PushSource(
    name="user_searches_stream",
    schema=input_schema,
    batch_config=trip_searches_function,
    description="User flight searches stream",
    owner="nlee@tecton.ai",
    tags={"release": "production"},
)

user_searches_eventsource.validate()
```

```python
from tecton import stream_feature_view, BatchTriggerType
from datetime import datetime, timedelta
from tecton.types import Field, Timestamp, Int64, String

@stream_feature_view(
    source=user_searches_eventsource,
    entities=[user],
    mode='python',
    online=True,
    offline=False,
    schema = input_schema,
    feature_start_time=datetime(2023,1,1),
    description='Searches from users on the platform',
    timestamp_field="transaction_timestamp",
    manual_trigger_backfill_end_time=datetime(2023,10,31),
    batch_trigger = BatchTriggerType.MANUAL,
    batch_schedule=timedelta(days=1)
)
def user_flight_search_fv(user_searches_eventsource):
    return user_searches_eventsource

user_flight_search_fv.validate()
```

```python
from tecton import RequestSource, on_demand_feature_view
from tecton.types import String, Timestamp, Int64, Field

request_schema = [Field('destination_airport', String)]

transaction_request = RequestSource(schema=request_schema)

@on_demand_feature_view(
    sources=[user_home_airport_fv, user_flight_search_fv, transaction_request],
    mode='python',
    description="Generates a flight booking link",
    schema = [Field('booking_url', String)]
)
def get_booking_url_v2(user_home_airport_fv, user_flight_search_fv, transaction_request):
    home_airport = user_home_airport_fv["home_airport_code"]
    destination_airport = transaction_request['destination_airport']
    departure_date = str(user_flight_search_fv['departure_date']).split()[0] #only get YYYY-MM-DD
    return_date = str(user_flight_search_fv['return_date']).split()[0]

    return {
        'booking_url': (f"https://www.kayak.com/flights/{home_airport}-{destination_airport}/{departure_date}/{return_date}?sort=bestflight_a")
    }
get_booking_url.validate()
```

```python
from tecton import FeatureService

get_booking_url_fs = FeatureService(
    name="booking_url_service:v1",
    features=[get_booking_url_v2]
)
get_booking_url_fs.validate()
```
